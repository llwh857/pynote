Python环境搭建(Windows)
---

必备软件：
---
- Python解释器

- Python编辑器


Windows 安装Python解释器
----
- 执行安装程序，路径可以选择默认或者自己挑选一个即可。
- 安装完成之后，我们需要设置系统的环境变量。


Windows环境变量：
---
环境变量是Windows系统的一个设置，输入一个程序的名称能够快速进行打开。 主要是在cmd(命令行中使用)。




检查python环境是否安装成功
---
- 当我们安装完成之后，我们打开cmd，输入python如果出现版本信息代表安装成功。
- 没有出现版本信息，请再次配置环境变量。

python包是什么？有什么用？
---



python包：
python包是一种插件，能够使我们处理某类数据更加方便快捷。
举个栗子：
-  鞋子：游戏中道具鞋子可以使人物移动更快
- 铠甲： 可以使人物抵御攻击，减少伤害。

游戏中的装备相当于一个个包，增加不同属性的装备，对应的增加某些能力。
python也有这种包，而且提供还提供了管理包的一些办法。


pip python包管理
---

- 安装包：
>pip install 包的名称 
>> pip install ipython
- 卸载包：
> pip uninstall 包的名称
>> pip uninstall ipython

Python虚拟环境
---
- 为什么要有虚拟环境：
  > 不同的项目可能会使用不同的python包，为了是降低包与包之间的干扰，保证纯粹的开发环境。所以衍生出来虚拟环境。一类项目一个环境，项目之间互不干扰。

- 安装虚拟环境：
 > 使用pip 安装虚拟环境使用的包。
  >>pip install virtualenvwrapper-win

- 设置文件夹和系统环境变量
 > 新建环境变量： WORKON_HOME和对应的文件夹。
 
 > 创建环境名称： mkvirtualenv 环境名称
 
 > 进入环境： workon 环境名称
 
 > 退出环境： deactivate
