python基础答疑解惑-什么是包?如何更快的获取包
------------

01 什么是python包
---
- 我们先从网络上查找一下相关解释。
1. 解释： 包是诸多模块的集合
> 
2. 查找了很多解释， 老外描述的**比较到位**，但对于刚刚接触编程的初学者来说还是晦涩难懂。 
> 
3. 举个很形象的例子。 特斯拉是一个汽车产物，由诸多系统构成 exp: 动力系统，自动驾驶系统。
 - 自动驾驶系统由： 摄像头， 传感器， 车载处理器， 显示设备等模块构成。
 - 链接， 特斯拉自动驾驶。https://www.tesla.cn/models
 
4. **所以我们可以得出， python程序就是好比特斯拉， 那对应的各种系统就是 python包， 每个包里面包含一个或者多个模块。每个模块执行一个或者多个任务。**

 

02 为什么说python包是python迅速发展的助推剂？
------
- 话题引入， 特斯拉的自动驾驶系统好比一个python包， 包里面有很多自动驾驶相关的模块。

- 话题延伸， 如果我需要图像识别，我就需要引入一个图像识别的包，用包里面自带的模块。 如果我需要做人工智能，同理， 引入人工智能包，帮助我们处理复杂计算。 

- 结论： 如果有各种各样的包，我们是不是直接可以调用？避免了我们写深奥的逻辑，学习难理解的线性变换等。 所以说，包的丰富，带来了开发的方便。





03 如何获取python包
------
1. 包存放在哪里？ 互联网上， 比如：国外的服务器上面，以及国内各大高校，社会组织的服务器上。
2. 如何获取并安装？ 比较复杂的就是下载并自行编译，安装， *另一种*是使用python提供的命令便捷安装 **pip**
