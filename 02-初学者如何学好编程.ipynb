编程三部曲
---
- 学习语言要**记忆**
- 学习语言要**理解**
- 好记性不如动手写**多练习**

重温： 
----

>计算机程序是由计算机语言编写组合而成的。

>计算机语言的作用是连接人类和电脑沟通的桥梁。

>学习计算机编程语言和学习汉语、英语的历程是一样的。

记忆：
----
    - 汉语需要记忆这个字是什么意思，写法是什么，词语之间应该怎么样搭配使用
    - 程序中也需要记忆一些特殊单词含义，用法。 print，input，debug。。。

理解：
---
    - 当我们掌握英语单词的意思之后，我们就要去尝试阅读英文长句子，短文章。去理解单词后面想要表达的意义。
    - 程序也是一样，掌握基本的写法之后，就要去理解代码为什么这样写？解决什么问题？修改之后会有什么变化？
    - 理解的意义在于读懂别人代码，理解别人的意图，当下次自己遇到类似的问题能够从中吸取经验，帮助自己解决问题。

起点：代码一W行：
---

*俗话说，光说不练假把式。那就先定一个小目标，积累代码1w行。*

学程序离不开写代码，有效的代码积累能够产生从量变到质变的过程。
这里我有一个要求：
    1. 1w行代码不算copy的。
    2. 1w行代码不算誊写的。

初学者都会犯的小毛病：
---

- 看代码不能理解 -- 这种情况一般是刚接触代码，英文环境陌生，手速慢，交流对象发生变化。
    >定要克服心理作用，坚持用记忆，理解，练习的方式去一遍遍的重复。也许重复几遍你就豁然开朗了。
-  对自己要求不严格， 代码看会和能写出来是两码事儿。只有当你能在编辑器中顺利的写出来才能证明你掌握了。
-  练习题打折扣。 一般情况下，练习题应该自己写四遍。
    >1. 照着誊写一遍，在誊写的过程中要去记忆自己不会的格式，去读懂代码的含义。
    
    >2. 尝试自己写一遍，在不看原来代码基础上，自己写，发现错误一定要去问、去搜，自己去解决。
    
    >3. 完全自己写一遍，写之前要在脑子里面想整个逻辑。 写出来的代码要做到运行无报错。
    
    >4. 尝试自己修改一些数据，自己猜测程序结果并用代码验证自己的猜测结果。
