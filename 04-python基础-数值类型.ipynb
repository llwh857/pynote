  ``` python
  #!/usr/bin/env python
  # -*- coding: utf-8 -*-
 
  ```
  
  **代码编写**：
  
  - 定义一个名字 a = 100
  
  
  - 定义一个名字 b = 200

  - 发现不管用/报错怎么办？ ->需要掌握输出语句 print(sum)
